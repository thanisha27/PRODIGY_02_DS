# Perform the data cleaning and EDA on the dataset for titanic.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_data = pd.read_csv("C://Users//Thanisha//Downloads//train.csv")
```

```python
# Data Cleaning

# Check for missing values
missing_values = titanic_data.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Impute missing values for Age and Embarked
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# Drop the Cabin column due to many missing values
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
# Check for missing values after imputation
print("\nMissing Values After Imputation:")
print(titanic_data.isnull().sum())
```

```python
# Ensure 'Sex' and 'Embarked' columns are encoded as numeric values (if not already)
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 0, 'female': 1})
titanic_data['Embarked'] = titanic_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Exclude non-numeric columns from the correlation matrix
numeric_columns = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex', 'Embarked']
correlation_matrix = titanic_data[numeric_columns].corr()
```

```python
# EDA - Exploratory Data Analysis

# Countplot of survivors
sns.countplot(data=titanic_data, x='Survived')
plt.title("Count of Survivors")
plt.show()
```

# Explore the relationship between variable and identify pattern and treads in the data

```python
# Relationship between Pclass and Survival
sns.countplot(data=titanic_data, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Relationship between Sex and Survival
sns.countplot(data=titanic_data, x='Sex', hue='Survived')
plt.title("Survival by Sex")
plt.show()
```

```python
# Age distribution
sns.histplot(data=titanic_data, x='Age', kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Fare distribution
sns.histplot(data=titanic_data, x='Fare', kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
# Relationship between SibSp (Number of Siblings/Spouses) and Survival
sns.countplot(data=titanic_data, x='SibSp', hue='Survived')
plt.title("Survival by SibSp")
plt.show()

```

```python
# Relationship between Parch (Number of Parents/Children) and Survival
sns.countplot(data=titanic_data, x='Parch', hue='Survived')
plt.title("Survival by Parch")
plt.show()
```

```python
# Embarked location and Survival
sns.countplot(data=titanic_data, x='Embarked', hue='Survived')
plt.title("Survival by Embarked Location")
plt.show()
```

```python
# Create the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
